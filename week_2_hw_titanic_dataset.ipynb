```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Titanic data set

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Seabron 

```python
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
```

```python
sns.barplot(x='class', y='survived', data=df)
plt.title('Survival Rate by Class')
```

```python
sns.violinplot(x='survived', y='age', data=df)
plt.title('Survival Rate by Age')
```

```python
sns.barplot(x="embark_town", y='survived', data=df)
plt.title('Survival Rate by Embark Town')
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(df['age'],kde=True)
plt.title('Age Distribution')
```

### Matplotlib

```python
plt.hist(df['age'], bins=30, edgecolor='black')
plt.title("Age Distribution")
```

```python
plt.bar(df["class"],df["fare"])
plt.title("Fare Distribution by Class")
```

```python
plt.imshow(df.corr())
plt.title("Correlation heat map")
```

### Pandas

```python
survival_rate_gender = df.groupby('sex')['survived'].mean()
survival_rate_age = df.groupby('age')['survived'].mean()
survival_rate_class = df.groupby('class')['survived'].mean()
```

```python
survival_rate_gender.plot(kind="bar")
```

```python
survival_rate_class.plot(kind="bar")
```
